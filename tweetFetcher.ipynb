```python
import tweepy
import pandas as pd
import datetime as datetime
```

```python
consumer_key = ""
consumer_secret = ""
access_token = ""
access_token_secret = ""
bearer_token = ""
```

```python
client = tweepy.Client(bearer_token=bearer_token, consumer_key=consumer_key, consumer_secret=consumer_secret, access_token=access_token, access_token_secret=access_token_secret, 
               wait_on_rate_limit=False)
```

```python
userid = "2360314352" 
response = client.get_users_tweets(id=userid, exclude="retweets", max_results = 100)
tweets = []
for data in response[0]:
    tweets.append(str(data))
i = 0
while (i<4):
    response = client.get_users_tweets(id=userid, exclude="retweets", max_results = 100, pagination_token = response[3]["next_token"])
    for data in response[0]:
        tweets.append(str(data))
    i += 1
```
